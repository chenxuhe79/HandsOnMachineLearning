## Chapter 8: Dimensionality Reduction
2022-12-25

```python
import time
import pandas as pd
import numpy as np

from sklearn.datasets import fetch_openml
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.metrics import mean_squared_error, accuracy_score, confusion_matrix, classification_report
from sklearn.ensemble import RandomForestClassifier
from sklearn.manifold import TSNE

from matplotlib import pyplot as plt
```

```python
mnist = fetch_openml('mnist_784',version =1)
```

```python
for k in mnist:
    print(k,end = ", ")
```

```python
X_mnist, Y_mnist = mnist['data'], mnist['target']
```

```python
pca = PCA(n_components = 154)
X_mnist_reduced =pca.fit_transform(X_mnist)
X_mnist_recovered = pca.inverse_transform(X_mnist_reduced)
```

```python
idxLoc = 36000
x=np.array(X_mnist.iloc[idxLoc]).reshape(28,28)
x_recovered = X_mnist_recovered[idxLoc,:].reshape(28,28)
_, (ax1, ax2) = plt.subplots(nrows = 1,ncols = 2, figsize =(3,6))
ax1.imshow(x, cmap='gray', interpolation='nearest', vmin=0, vmax=256)
ax2.imshow(x_recovered, cmap='gray', interpolation='nearest', vmin=0, vmax=256)
plt.show()
```

```python
X_train, X_test = X_mnist.iloc[:60000], X_mnist.iloc[60000:]
Y_train, Y_test = Y_mnist[:60000], Y_mnist[60000:]
```

```python
rf_clf = RandomForestClassifier(n_estimators = 100, random_state =42)
t0 = time.time()
rf_clf.fit(X_train, Y_train)
t1 =time.time()
print(f'The training time is {t1-t0:.2f}s')
```

```python
Y_predict = rf_clf.predict(X_test)
```

```python
print(classification_report(Y_test,Y_predict, digits = 4))
```

#### Comparison with reduced data

```python
pca = PCA(n_components = 0.95)
X_train_reduced = pca.fit_transform(X_train)
X_test_reduced = pca.transform(X_test)
```

```python
rf_clf2 = RandomForestClassifier(n_estimators = 100, random_state =42)
t0 = time.time()
rf_clf2.fit(X_train_reduced, Y_train)
t1 =time.time()
print(f'The training time is {t1-t0:.2f}s')
```

```python
print(classification_report(Y_test,rf_clf2.predict(X_test_reduced),digits = 4))
```

#### Visualization

```python
np.random.seed(42)
m = 10000
idx = np.random.permutation(60000)[:m]

X_Vis = mnist['data'].iloc[idx]
y_Vis = np.array(mnist['target'][idx]).astype(np.float)
```

```python
tsne = TSNE(n_components=2, random_state=42)
X_reduced = tsne.fit_transform(X_Vis)
```

```python
plt.figure(figsize = (12,10))
plt.scatter(X_reduced[:,0],X_reduced[:,1], c=y_Vis, cmap = 'jet')
plt.colorbar()
plt.show()
```
