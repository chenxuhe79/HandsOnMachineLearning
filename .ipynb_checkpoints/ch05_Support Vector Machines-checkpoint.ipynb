## 1. Linear SVM classification

```python
import numpy as np
from sklearn import datasets
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler 
from sklearn.svm import LinearSVC
from matplotlib import pyplot as plt
from matplotlib.colors import ListedColormap
```

#### Iris data set

```python
iris = datasets.load_iris()
X = iris["data"][:, (2, 3)] # petal length, petal width
y = (iris["target"] == 2).astype(np.float64) # Iris-Virginica
svm_clf1 = Pipeline([
("scaler", StandardScaler()),
("linear_svc", LinearSVC(C=1, loss="hinge")), ])
svm_clf1.fit(X, y)
w1 = 1/np.sqrt(np.sum(svm_clf1['linear_svc'].coef_[0]**2))

svm_clf2 = Pipeline([
("scaler", StandardScaler()),
("linear_svc", LinearSVC(C=100, loss="hinge")), ])
svm_clf2.fit(X, y)
w2 = 1/np.sqrt(np.sum(svm_clf2['linear_svc'].coef_[0]**2))
```

```python
X0=X[y==0]
X1=X[y==1]

xlist = np.linspace(1, 7, 100) # Create 1-D arrays for x,y dimensions
ylist = np.linspace(0, 2.5,100) 
Xlist,Ylist = np.meshgrid(xlist, ylist) # Create 2-D grid xlist,ylist values
x_g, y_g = Xlist.flatten(), Ylist.flatten()
x_g, y_g = x_g.reshape((len(x_g), 1)), y_g.reshape((len(y_g), 1))
grid = np.hstack((x_g, y_g))
```

```python
plt.scatter(X0[:,0],X0[:,1],marker = "s")
plt.scatter(X1[:,0],X1[:,1],marker = "^")
plt.xlabel("petal length(cm)")
plt.ylabel("petal width(cm)")
plt.title("C=100: more margin violations with width = "+str(round(w2,2)))
plt.contourf(Xlist, Ylist, svm_clf2.predict(grid).reshape(Xlist.shape), alpha = 0.25, cmap = ListedColormap(("red","blue")))
```

```python
plt.scatter(X0[:,0],X0[:,1],marker = "s")
plt.scatter(X1[:,0],X1[:,1],marker = "^")
plt.xlabel("petal length(cm)")
plt.ylabel("petal width(cm)")
plt.title("C=1: less margin voilations with width = "+str(round(w1,2)))
plt.contourf(Xlist, Ylist, svm_clf1.predict(grid).reshape(Xlist.shape), alpha = 0.25, cmap = ListedColormap(("red","blue")))
```

#### Moon's data set

```python
import seaborn as sns
from sklearn.datasets import make_moons
from sklearn.preprocessing import PolynomialFeatures

X, y = make_moons(noise=0.32, random_state=1, n_samples=100)
sns.scatterplot(
    x=X[:, 0], y=X[:, 1], hue=y, marker="o", s=25, edgecolor="k", legend=False
).set_title("Data")
```

```python
poly_svm_clf = Pipeline([
    ("poly_features", PolynomialFeatures(degree=3)), ("scaler", StandardScaler()),
    ("svm_clf", LinearSVC(C=10, loss="hinge"))]) 
poly_svm_clf.fit(X, y)
```

```python
X0=X[y==0]
X1=X[y==1]

xlist = np.linspace(-2, 3, 100) # Create 1-D arrays for x,y dimensions
ylist = np.linspace(-1.5, 1.8,100) 
Xlist,Ylist = np.meshgrid(xlist, ylist) # Create 2-D grid xlist,ylist values
x_g, y_g = Xlist.flatten(), Ylist.flatten()
x_g, y_g = x_g.reshape((len(x_g), 1)), y_g.reshape((len(y_g), 1))
grid = np.hstack((x_g, y_g))
```

```python
#plt.scatter(X0[:,0],X0[:,1],marker = "s")
#plt.scatter(X1[:,0],X1[:,1],marker = "^")
sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=y, marker="o", s=30, edgecolor="k", legend=False)
plt.contourf(Xlist, Ylist, poly_svm_clf.predict(grid).reshape(Xlist.shape), alpha = 0.25, cmap = ListedColormap(("red","blue")))
```

```python
from sklearn.svm import SVC
polykernel_svm_clf = Pipeline([
    ("scaler", StandardScaler()),
    ("svm_clf", SVC(kernel = 'poly', degree = 3, C=5, coef0=1))]) 
polykernel_svm_clf.fit(X, y)
```

```python
sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=y, marker="o", s=30, edgecolor="k", legend=False)
plt.contourf(Xlist, Ylist, polykernel_svm_clf.predict(grid).reshape(Xlist.shape), alpha = 0.25, cmap = ListedColormap(("red","blue")))
```

```python
from sklearn.svm import SVC
rbfkernel_svm_clf = Pipeline([
    ("scaler", StandardScaler()),
    ("svm_clf", SVC(kernel = 'rbf', gamma = 0.01, C=1000))]) 
rbfkernel_svm_clf.fit(X, y)
```

```python
sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=y, marker="o", s=30, edgecolor="k", legend=False)
plt.contourf(Xlist, Ylist, rbfkernel_svm_clf.predict(grid).reshape(Xlist.shape), alpha = 0.25, cmap = ListedColormap(("red","blue")))
```
